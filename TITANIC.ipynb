# Data Cleaning

Remove Null values

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Pclass",y="Age",data=titanic)
sns.set_style("darkgrid")
plt.show()
```

```python
def imput_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
titanic.isnull().sum()
```

```python
titanic["Age"]=titanic[["Age","Pclass"]].apply(imput_age,axis=1)
```
