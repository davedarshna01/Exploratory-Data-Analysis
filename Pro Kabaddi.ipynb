```python
# Plot for Total Players by their Position 

fig, axx =plt.subplots(figsize=(12,6))
axx.set_title("Total Players by their Position",size=20,color="black")
sns.barplot(x='Total_player', y='Position',hue='Position', data=Player, palette='Set1')
```

```python
#find position of the player

A=['All Rounder','Defender','Raider']
B=Player_data.loc[Player_data["Position"].isin(A)]
B


#find total number of the position

C=B.groupby(['Position'],as_index=False).count()[['Position','Name']]
C.columns=["Position","Total player"]


#find total position count of each type

D=B.groupby(['Position','Name'],as_index=False).count()[['Position']].count()
D.columns=['Name','Position']

C
```
